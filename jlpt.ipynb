```python
#certified rate difference between july and december (run ttest to check real difference between jul and dec exams)
#by exam level and total is mean of all level by years
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

july = pd.read_csv('jlpt - july.csv')
july.columns = ['Year', 'N1', 'N2', 'N3', 'N4', 'N5', 'Total']
july = july.loc[1:]
dec = pd.read_csv('jlpt - dec.csv')
dec.columns = ['Year', 'N1', 'N2', 'N3', 'N4', 'N5', 'Total']
dec = dec.loc[1:]
t_list = []
p_list = []
for level in range(1, 6):
    t, p = ttest_ind(july['N' + str(level)], dec['N' + str(level)])
    t_list.append(t)
    p_list.append(p)
    
t, p = ttest_ind(july['Total'], dec['Total'])
t_list.append(t)
p_list.append(p)
t_Series = pd.Series(t_list, index= ['N1', 'N2', 'N3', 'N4', 'N5', 'Total']).rename('t_value')
p_Series = pd.Series(p_list, index= ['N1', 'N2', 'N3', 'N4', 'N5', 'Total']).rename('p_value')

ttest_result = pd.DataFrame([t_list, p_list]).T
ttest_result.index = ['N1', 'N2', 'N3', 'N4', 'N5', 'Total']
ttest_result.columns = ['t_value', 'p_value']
ttest_result
```
